## Build Your Own Statistics Library

Without importing anything (including  math, statistics, numpy, etc.), complete the empty functions to return the following:

- Square root
- Variance (population not sample - see - stretch goals)
- Standard deviation (again, population not sample)
- Z-Score
- Outliers with a z-score above a threshold argument

```
# square root function.
def sqrt(n):
  return n ** 0.5
# varaince function.
def variance(nums):
  mean = sum(nums) / len(nums)
  return sum((x - mean) ** 2 for x in nums) / len(nums)
# standad deviation function.
def stdev(nums):
  return sqrt(variance(nums))
# z score function.
def zscore(n, nums):
  mean = sum(nums) / len(nums)
  return (n - mean) / stdev(nums)
# outliers function.
def find_outliers(nums, threshold=1):
  outliers = []
  for i in range(len(nums)):
    if abs(zscore(nums[i],nums)) > threshold:
      outliers.append(nums[i])
  return outliers
```

```
# driver program.

# set n.
n = 12
# set nums.
nums = 1, 6, 13, 24, 39, 55, 72
# show square root.
print ('squae root of n:', sqrt(n))
# show variance.
print('variance of nums:', sqrt(n))
# show standard deviation.
print('standard deviation of nums:', sqrt(n))
# show z score. 
print('z score of n & nums:', sqrt(n))
# show outliers.
print('outliers of nums:', sqrt(n))
```
