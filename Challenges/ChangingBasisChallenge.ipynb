Write a function that takes a decimal number and base as input and outputs a string representation of that number in that base. For example, 
decimal_to_base(num=5, base=2) == '101'
decimal_to_base(45,16) == '2D'

We're just concerned with the number itself, so don't bother with header representations like '0b' or '0x' if you're familiar with those. If you're unfamiliar with numbers in bases > 10, look up how hexadecimal numbers work.


Stretch goal: Write a reverse function base_to_decimal(num, base) that takes in a number and the base its in and outputs its decimal representation.

Stretch goal 2: Write a function base_to_base(num, base1, base2) that takes a number in an arbitrary base and outputs its representation in an arbitrary base. Hint: There may be a clever way of using the functions you've already written.


```
# this string may be useful.
#arbitrary length, currently base 36.

# decimal to base function.
def decimal_to_base(num, base):
  answer = ''
  while num > 0:
    answer = base_units[num % base] + answer
    num = num // base
  return answer
# base to decimal function.
def base_to_decimal(num, base):
  num = str(num)
  answer = 0
  for n in num:
    answer = base * (answer + base_units[:base].index(n))
  return answer
# base to base function.
def base_to_base(num, start_base, end_base):
  return decimal_to_base(base_to_decimal(num, start_base), end_base)
```

```
# driver program.

# given base units.
base_units='0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
# show decimal to base with example.
print(decimal_to_base(num=5, base=2))
print(decimal_to_base(45,16))
```
