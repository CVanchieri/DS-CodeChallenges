Puzzle 1:
The function evenOddNoIfs(n) takes an integer and returns "odd" of the integer is "odd" and "even" if it is "even". The puzzle part is that you can not use any if statements or other conditionals! (this is honor system and not tested for, only the correct output is tested)

```
# even odd function.
def evenOddNoIfs(n):
  status = ['even', 'odd']
  return status[int(n)%2]
```

```
# driver program.

# show even odd result.
evenOddNoIfs(43)
```

Puzzle 2:
The function exclaim(s) takes a string and returns that string with a normal exclamation point on the end. The puzzle part is that you can not use an exclamation point in your function! (again, honor system)

```
# exclaim function.
def exclaim(s):
  return f"{s}{bytes([33]).decode('utf-8')}"
```

```
# driver program.

# show exclaim result.
exclaim('you did it')
```

Puzzle 3:
The function equal(x, y) takes two numbers and returns "equal" if they are "equal" and "not equal" if they are not equal. The puzzle part is that you can not arithmetic operators (+, -, *, **, /, //, %) or comparison operators (==, !=, <, <=, >, >=)! (hey there still!)

```
# equal function.
def equal(x, y):
  verify = ['Not Equal', 'Equal']
  return verify[x is y]
```

```
# driver program.

# show equal result.
equal(12, 12)
```
