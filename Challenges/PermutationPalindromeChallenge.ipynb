Write a function that, given a string, determines whether any permutation of that string is a palindrome. 

Examples:

- "civic" should return True
- "ivicc" should return True
- "civil" should return False
- "livci" should return False

Yes, yes, "ivicc" itself is not a palindrome, but it can be rearranged into a valid palindrome, i.e. "icvci", which is a valid palindrome. 

```
# imports.
from collections import Counter

# set # of characters.
NO_OF_CHARS = 256
# permutation palindrome function.
def is_permutation_palindrome(string):
  
    # create a count array and initialize.   
    # all values as 0. 
    count = [0] * (NO_OF_CHARS) 
  
    # for each character in input strings, increment count in the corresponding count array.
    for i in range( 0, len(string)) : 
        count[ord(string[i])] = count[ord(string[i])] + 1
  
    # count odd occurring characters. 
    odd = 0
      
    for i in range(0, NO_OF_CHARS ) : 
        if (count[i] & 1) : 
            odd = odd + 1
  
        if (odd > 1) : 
            return False
              
    # return true if odd count is 0 or 1.  
    return True
  
```

```
# driver program.

# show results for civic.
if(is_permutation_palindrome("civic")) : 
    print("True") 
else : 
    print("False") 
# show results for ivicc.      
if(is_permutation_palindrome("ivicc")) : 
    print("True") 
else : 
    print("False")
# show results for civil.
if(is_permutation_palindrome("civil")) : 
    print("True") 
else : 
    print("False") 
# show results for livci.      
if(is_permutation_palindrome("livci")) : 
    print("True") 
else : 
    print("False")
```
