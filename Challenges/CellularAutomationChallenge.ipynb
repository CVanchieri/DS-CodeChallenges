There are 8 cells in a row, each cell contains a 1 or 0. At each step, the contents of the cells change according to the following rules:
If a cell has 2 adjacent neighbors that are both 1 or both 0, the cell becomes 1.
Otherwise it becomes 0.
Given the initial state, return the state of the system after N steps.
```
#Example:
Input: cells = [0,1,0,1,1,0,0,1], N==7

Output: [0,0,1,1,0,0,0,0]
```

````
# cells after n steps function.
def cells_after_N_steps(cells, N):
  # nextday function.
  def nextday(cells):
    return [int(i>0 and i<7 and cells[i-1] == cells[i+1]) for i in range(8)]
  seen = {}
  while N>0:
    c = tuple(cells)
    if c in seen:
      N %= seen[c] - N 
    seen[c] = N
    if N >= 1:
      N -= 1 
      cells = nextday(cells)
  return cells
````

````
# driver program.

# set inputs.
cells = [0,1,0,1,1,0,0,1]
# set N.
N = 7
# show rsults for calls after N steps.
cells_after_N_steps(cells, N)
````
